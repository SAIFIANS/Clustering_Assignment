#Hierarchy Clustering

```python
#importing the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
#Read the dataset
df = pd.read_csv("crime_data.csv")
df.head()
```

```python
#normalizing the data
df_norm = normalize(df.iloc[:,1:])
df_norm
```

```python
#plotting a dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method ='single'))
```

```python
#performing a Hierarchical clustering
hc =AgglomerativeClustering(n_clusters=4,
                            metric="euclidean",
                            linkage= 'single')
```

```python
#predicting the data using the model
y_hc = hc.fit_predict(df_norm)
Clusters = pd.DataFrame(y_hc, columns= ['Clusters'])
```

```python
y_hc
```

```python
#Storing the Cluster Id into the dataframe
df["Hierarchy_ClusterId"] = Clusters
df
```

#KMeans Clustering

```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
#reading the dataset
crime = pd.read_csv("crime_data.csv")
crime.head()
```

```python
#dropping the Unnamed column
crime.drop(["Unnamed: 0"], inplace=True, axis=1)
crime.head()
```

```python
#Standardizing the data
stan_df = StandardScaler().fit_transform(crime)
stan_df
```

```python
#plotting an elbow chart to identify the suitable no. of clusters
wcss= []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, random_state=0)
  kmeans.fit(stan_df)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title("Elbow Chart")
plt.xlabel("No. of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
#Building the model using KMeans
kmeans=KMeans(n_clusters=4, random_state= 12)
kmeans.fit(stan_df)
```

```python
#Storing the clusters into the dataframe
crime["KMeans_Clusters"] = kmeans.labels_
crime
```

```python
#plotting a scatter plot using Murder attribute on x-axis, KMeans_Clustering Id on y-axis
plt.figure(figsize=(11,8))
plt.scatter(crime["Murder"], crime["KMeans_Clusters"], c=kmeans.labels_)
```

#DBSCAN Clustering

```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
```

```python
#reading the dataset
crime = pd.read_csv("crime_data.csv")
crime.head()
```

```python
#Dropping the Unnamed column
crime.drop(["Unnamed: 0"], inplace=True, axis=1)
crime
```

```python
#Standardizing the values using StandardScaler
std_df = StandardScaler().fit_transform(crime.values)
std_df
```

```python
#Building the model using K-Nearest Neighbors
nbrs = NearestNeighbors(n_neighbors=5).fit(std_df)
```

```python
# Storing the values in dist and indices variables
dist, indices = nbrs.kneighbors(std_df)
```

```python
dist = np.sort(dist, axis=0)
dist = dist[:,1]
```

```python
#plotting the dist variable
plt.plot(dist)
```

```python
#Building the model using DBSCAN algo. with epsilon values as 1.25 and minimun samples of 5
dbscan = DBSCAN(eps=1.25, min_samples=5)
dbscan.fit(std_df)
```

```python
#Storing the cluster ids into the dataframe
crime["DBSCAN_Clusters"] = dbscan.labels_
crime
```

```python
#Plotting a scatter plot with Murder attribute on x-axis and Clusterid created by DBSCAN on y-axis
plt.figure(figsize=(11,8))
plt.scatter(crime["Murder"], crime["DBSCAN_Clusters"], c= dbscan.labels_)
```
